```python
def key_deter_WC(cx,cy):
            if cy>150 and cy<300:
                if cx>20 and cx<220:
                    pyautogui.press("w")
                    return "w"
                elif cx>220 and cx<420:
                    pyautogui.press("c")
                    return "c"
                elif cx>420 and cx<620:
                    pyautogui.press("t")
                    return "t"
                else:
                    return ""
                
            else:
                return ""
            
```

```python
def getString_WC():
    
    cap = cv2.VideoCapture(0)
    count = 0

    result = ""
    while(1):

        ## Read the image
        ret, frame = cap.read()
        ## Do the processing
        frame = cv2.flip(frame,1)


        global cx
        global cy
        global old_area,new_area
        old_area,new_area=0,0
        #for yellow color idenitfiaction in frame
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        lower_yellow = np.array([14,141,140])#change this hsv values if yellow color as per your lighting condition
        upper_yellow = np.array([84,255,255])#same as above    

        mask = cv2.inRange(hsv, lower_yellow, upper_yellow)
        blur = cv2.medianBlur(mask, 15)
        blur = cv2.GaussianBlur(blur , (5,5), 0)
        thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)


        cv2.imshow("mask", mask)
        #find contours in frame
        contours, hierarchy = cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

        #function to determine which key is pressed based on the center of the contour(yellow paper)

        cv2.drawContours(frame,contours,-1,(0,255,0),2)

        if len(contours)>0:
            cnt=max(contours,key=cv2.contourArea)
            if (cv2.contourArea(cnt)>600 and cv2.contourArea(cnt)<1200):


                M = cv2.moments(cnt)

                cx = int(M['m10']/M['m00'])
                cy = int(M['m01']/M['m00'])
                 #print ("Centroid = ", cx, ", ", cy)
                new_area=cv2.contourArea(cnt)
                 #print("new area ",new_area)
                cv2.circle(frame,(cx,cy),1,(0,0,255),2)
                if count==0:
                    old_area=new_area
                     #print("in count==0   ",count)

                count=count+1
                 #print(count)
                if count==20:
                    count=0
                    diff_area=new_area-old_area
                    if diff_area>500 and diff_area<1200:
                        #print("diff- ",diff_area)
                        subs= key_deter_WC(cx,cy)
                        if(subs == ""):
                            pass
                        else:
                            #print(subs)
                            result += subs
                            break
                        cv2.putText(frame, "Enter name: " + result, (50,50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0),2)



        #display the keyboard in the screen        


        keyboard_layout_WC(frame)
        cv2.imshow('image',frame)
        if cv2.waitKey(1) == 27:  ## 27 - ASCII for escape key
            break
    ############################################

    ############################################
    ## Close and exit
    cap.release()
    cv2.destroyAllWindows()
    return result
    ############################################
```

```python
def key_deter_num(cx,cy):
            if cy>50 and cy<410 and cx > 50 and cx < 530:
                #first row  1,2,3,4
                if cy > 50 and cy < 170:
                    # 1
                    if cx > 50 and cx < 170:
                        pyautogui.press("1")
                        return "1"
                    # 2
                    elif cx > 170 and cx < 290:
                        pyautogui.press("2")
                        return "2"
                    # 3    
                    elif cx > 270 and cx < 410:
                        pyautogui.press("3")
                        return "3"
                    # 4
                    else:
                        pyautogui.press("4")
                        return "4"
                    
                    
                #second row 5,6,7,backspace
                
                elif cy > 170 and cy < 290:
                    # 5
                    if cx > 50 and cx < 170:
                        pyautogui.press("5")
                        return "5"
                        
                    # 6
                    elif cx > 170 and cx < 290:
                        pyautogui.press("6")
                        return "6"
                        
                    # 7    
                    elif cx > 290 and cx < 410:
                        pyautogui.press("7")
                        return "7"
                    
                    # Backspace
                    else:
                        pyautogui.press("backspace")
                        return "B"
                    
                
                #third row 8,9,0, Enter
                elif cy > 290 and cy < 410:
                    
                    # 8
                    if cx > 50 and cx < 170:
                        pyautogui.press("8")
                        return "8"
                        
                    # 9
                    elif cx > 170 and cx < 290:
                        pyautogui.press("9")
                        return "9"
                        
                    # 0    
                    elif cx > 290 and cx < 410:
                        pyautogui.press("0")
                        return "0"
                    
                    # Enter
                    else:
                        pyautogui.press("enter")
                        return "E"
                
                else:
                    return ""
                
            else:
                return ""
                
                
                
#                 if cx>20 and cx<220:
#                     print("w")
#                     pyautogui.press("w")
#                     return "w"
#                 elif cx>220 and cx<420:
#                     print("c")
#                     pyautogui.press("c")
#                     return "c"
#                 elif cx>420 and cx<620:
#                     print("t")
#                     pyautogui.press("t")
#                     return "t"
#                 else:
#                     return ""
                
#             else:
#                 return ""
            
```

```python
def getConfirmation(name, amount, balance):
    
    cap = cv2.VideoCapture(0)
    count = 0

    result = ""
    while(1):

        ## Read the image
        ret, frame = cap.read()
        ## Do the processing
        frame = cv2.flip(frame,1)
    
        message1 = "Hello " + name + "!, " + "You are about"
        message2 = "to withdraw " + str(amount) + " out of " + str(balance)
        
        cv2.putText(frame,  message1, (50,100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255),3)
        cv2.putText(frame,  message2, (50,130), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255),3)
        cv2.putText(frame, "Confirm", (120, 340), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255),3)
        cv2.putText(frame, "Cancel", (400, 340), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255),3)
        
        #cv2.putText(frame, "You entered: " + result, (50,440), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0),1)


        global cx
        global cy
        global old_area,new_area
        old_area,new_area=0,0
        #for yellow color idenitfiaction in frame
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        lower_yellow = np.array([14,141,140])#change this hsv values if yellow color as per your lighting condition
        upper_yellow = np.array([84,255,255])#same as above    

        mask = cv2.inRange(hsv, lower_yellow, upper_yellow)
        blur = cv2.medianBlur(mask, 15)
        blur = cv2.GaussianBlur(blur , (5,5), 0)
        thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)


        cv2.imshow("mask", mask)
        #find contours in frame
        contours, hierarchy = cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

        #function to determine which key is pressed based on the center of the contour(yellow paper)

        cv2.drawContours(frame,contours,-1,(0,255,0),2)

        if len(contours)>0:
            cnt=max(contours,key=cv2.contourArea)
            if (cv2.contourArea(cnt)>600 and cv2.contourArea(cnt)<1200):


                M = cv2.moments(cnt)

                cx = int(M['m10']/M['m00'])
                cy = int(M['m01']/M['m00'])
                 #print ("Centroid = ", cx, ", ", cy)
                new_area=cv2.contourArea(cnt)
                 #print("new area ",new_area)
                cv2.circle(frame,(cx,cy),1,(0,0,255),2)
                if count==0:
                    old_area=new_area
                     #print("in count==0   ",count)

                count=count+1
                 #print(count)
                if count==20:
                    count=0
                    diff_area=new_area-old_area
                    if diff_area>500 and diff_area<1200:
                        #print("diff- ",diff_area)
                        subs= key_deter_confirmation(cx,cy)
                        if(subs == ""):
                            pass
                        else:
                            #print(subs)
                            result += subs
                            break


        #display the keyboard in the screen        


        keyboard_layout_confirmation(frame)
        cv2.imshow('image',frame)
        if cv2.waitKey(1) == 27:  ## 27 - ASCII for escape key
            break
    ############################################

    ############################################
    ## Close and exit
    cap.release()
    cv2.destroyAllWindows()
    return result
    ############################################
```
